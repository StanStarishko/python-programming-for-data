### Exercise 1 - Line plot of net migration
---

Creating a line plot of mean net migration over the years 2015 to 2019

* create a new dataframe containing only the five columns holding net migration
* create a new data variable to hold the means of the five columns
* create a labels variable to hold the keys (column headings)
* use ```chart = sns.lineplot(data=prepared dataframe)``` to plot your line chart

***Presentation tips:***   
Plot the chart, then add formatting   

Rotate the x-axis labels in the plot:  
` chart.set_xticklabels(labels, rotation=30)`  

Show the grid:  
`sns.set_style("whitegrid")`

````python
import pandas as pd
import seaborn as sns


def is_valid_link(link="",link_name="",autotest=False):
  # link always isn't empty and must have is string
  return_value = link != "" and isinstance(link, str)

  if not return_value and not autotest: # not print if autotest
    print(f"{link_name} is not valid")

  return return_value


# get data from Excel
def get_excel_data(url="",sheet_name="default"):
  # url and sheet name always isn't empty and must have is string

  is_not_valid_url = not is_valid_link(url,"url")
  is_not_valid_sheet_name = not is_valid_link(sheet_name,"sheet name")

  if is_not_valid_url or is_not_valid_sheet_name:
    return False

  if sheet_name == "default":
    df = pd.read_excel(url)
  else:
    df = pd.read_excel(url,sheet_name)

  return df


# get data from CVS
def get_csv_data(url):

  if not is_valid_link(url,"url"):
    return False

  df = pd.read_csv(url)
  return df


# Line plot of net migration
def show_lineplot(df):
  # Creating a line plot of mean net migration over the years 2015 to 2019
  # create a new dataframe containing only the five columns holding net migration
  labels = ["2015","2016","2017","2018","2019"]

  # create a new data variable to hold the means of the five columns
  filtered_columns = filtered_columns = df[[col for col in df.columns if any(label in col for label in labels)]]
  numerical_df = filtered_columns.select_dtypes(include="number")
  data = numerical_df.mean()

  # create a labels variable to hold the keys (column headings)
  labels = list(numerical_df.columns)

  # use chart = sns.lineplot(data=prepared dataframe) to plot your line chart
  chart = sns.lineplot(x=labels, y=data)

  # Rotate the x-axis labels in the plot:
  chart.set_xticklabels(labels, rotation=30)

  # Show the grid:
  sns.set_style("whitegrid")


# execute procedure from main context
url = "https://github.com/futureCodersSE/working-with-data/blob/main/Data%20sets/public_use-talent-migration.xlsx?raw=true"
df = get_excel_data(url, "Country Migration")
show_lineplot(df)

````

### Exercise 2 - Creating a Bar chart for yearly migration
---
Create a bar chart which shows the total net migration for the years 2015-2019  

* split the migration columns into a new dataframe
* create a data variable, for the y values, from the max() of the five columns
* create a labels variable, this time just create a list of the 5 years ['2015','2016','2017','2018','2019']
* plot the bar chart (`sns.barplot(x=labels, y=y values)` )

***Presentation tips***:
* use `chart.set_xlabel('Year')` and `chart.set_ylabel('Maximum net migration')` to name your axis  

````python
# Creating a Bar chart for yearly migration
def show_barplot(df):
  # split the migration columns into a new dataframe
  labels = ["2015","2016","2017","2018","2019"]

  # create a new data variable to hold the means of the five columns
  filtered_columns = filtered_columns = df[[col for col in df.columns if any(label in col for label in labels)]]
  numerical_df = filtered_columns.select_dtypes(include="number")
  data = numerical_df.max()

  # plot the bar chart (`sns.barplot(x=labels, y=y values)` )
  chart = sns.barplot(x=labels, y=data)
  chart.set_xlabel('Year')
  chart.set_ylabel('Maximum net migration')


# execute procedure from main context
show_barplot(df)
````

### Exercise 3 - creating a bar graph using grouped data
---

Create a horizontal bar chart of 2019 mean immigration for each income level ('`target_country_wb_income`')

* create a data variable which contains the means, grouped by '`target_country_wb_income`'
* extract your labels (x) using the .keys() function
* use the `sns.barplot` with to create a horizontal bar graph (*Hint: swap the axes so labels is the y axis*)
* add labels to the axes ('Year', 'Net Migration')  
* show the plot  

Try plotting as a vertical bar chart - can you see why horizontally is more appropriate?

````python
import matplotlib.pyplot as plt

# creating a bar graph using grouped data
def show_grouped_barplot(df):
  # create a data variable which contains the means, grouped by 'target_country_wb_income'
  numerical_df = df.filter(like="net_per_10K_2019")
  regions = df.groupby("target_country_wb_income")[numerical_df.columns].mean()
  data = regions["net_per_10K_2019"]

  # extract your labels (x) using the .keys() function
  labels = data.keys()

  # use the sns.barplot with to create a horizontal bar graph (Hint: swap the axes so labels is the y axis)
  # plotting as a vertical bar chart
  plt.figure(figsize=(10, 6))
  sns.set_style("whitegrid")
  chart = sns.barplot(x=labels, y=data)
  chart.set_title('Vertical Bar Plot of Net Migration')
  chart.set_xlabel('Income Level')
  chart.set_ylabel('Net Migration')

  plt.show()
  print()

  # plotting as a horizontal bar chart
  plt.figure(figsize=(10, 6))
  sns.set_style("whitegrid")
  chart = sns.barplot(x=data, y=labels)
  chart.set_title('Horizontal Bar Plot of Net Migration')
  chart.set_xlabel('Net Migration')
  chart.set_ylabel('Income Level')

  plt.show()

# execute procedure from main context
show_grouped_barplot(df)
````

### Exercise 4 - Creating a pie chart of stress data
---
Create a pie chart which shows how stressed students who speak different languages are.   

To do this:

* similar to Exercise 2 - create a variable which groups the means of data by firstlanguage  
* store the means for 'Stress' in a variable called stress_data
* extract your labels using the keys() function

Seaborn doesn't have a function for plotting pie charts but you can use Seaborn functions for styling pie charts created by matplotlib.

* add an import statement above your funciton to import the matplotlib.pyplot library, aliased as plt
* use the Seaborn function `colors = sns.color_palette('pastel')`to create a colour palette for the chart.  (_Hint: you can find a list of available palettes [here](https://seaborn.pydata.org/tutorial/color_palettes.html)_)
* plot your pie chart using `plt.pie()` adding parameters to set labels and a color theme **colors = colors**
* write a comment noting anything interesting about the visualisation




````python
#Creating a pie chart of stress data
def show_piechart(df):
  # Grouping by language and calculating the average stress value for each language
  stress_data = df.groupby('firstlanguage')['Stress'].mean()

  # Create labela
  labels = stress_data.index

  # Set the colour palette
  colors = sns.color_palette('muted')

  # Creating a pie chart
  plt.figure(figsize=(10, 6))
  plt.pie(stress_data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
  plt.title('Stress Levels by Language')

  plt.show()



# execute procedure from main context
url = "https://raw.githubusercontent.com/lilaceri/Working-with-data-/b157a2feceb7709cf82426932385706d65446270/Data%20Sets%20for%20code%20divisio/Positive_Psychology_2017.csv"
df = get_csv_data(url)
show_piechart(df)
````

### Exercise 5 - Creating a box plot of Wellbeing
---
A box plot is used to visualise summary infomation about a data series such as the min, max and median.

Create a box plot of the Wellbeing scores

*  split off the wellbeing column into a new dataframe
*  create a label list containing the label ['Wellbeing']
*  use `chart = sns.boxplot(data=new_df)` to create a boxplot
*  set the x-axis label using `chart.set_xticklabels(labels)`

````python
import numpy as np

# Creating a box plot of Wellbeing
def show_boxplot(df):
  # Extracting a Wellbeing column into a new DataFrame
  new_df = df[['Wellbeing']]

  # Creating labels
  labels = ['Wellbeing']

  # Creating a boxplot
  plt.figure(figsize=(6, 10))
  chart = sns.boxplot(data=new_df)
  chart.set_xticklabels(labels)
  plt.title('Box Plot of Wellbeing Scores')

  # Setting the Y-axis steps
  # Create a list of values from 10 to 40 with a step of 5
  y_ticks = np.arange(10, 40, 5)

  # Set the y-axis ticks to the list of values
  chart.set_yticks(y_ticks)

  plt.show()



# execute procedure from main context
show_boxplot(df)
````

### Exercise 7 - Create a scatterplot of Wellbeing and Stress with line of best fit
---

Assuming that Stress is fairly closely associated with Wellbeing:

Create a scatterplot of Wellbeing and Stress data.

* create **x** from the `Stress` column
* create **y** from the `Wellbeing` column
* use `chart=sns.scatterplot(x=x,y=y)` to create a scatterplot
* add x axis and y axis labels using `chart.set_xlabel('Stress')` and `chart.set_ylabel('Wellbeing')`

Adding a line of best fit:   
* the Seaborn library has a function that will plot a scatter plot with a line of best fit generated from a linear regression
* replace the instruction to create the scatter plot with `chart=sns.regplot(x=x, y=y)`

Write a short data story ( a description of the data.  What conclusion could be reached from the chart?  How confident could you be in this conclusion and why?




````python
# Create a scatterplot of Wellbeing and Stress with line of best fit
def show_scatterplot(df):
  # Create x and y variables
  data_x = df['Stress']
  data_y = df['Wellbeing']

  # Create scatterplot with a line of best fit
  plt.figure(figsize=(10, 6))
  chart = sns.regplot(x=data_x, y=data_y,
                      scatter_kws={'color': 'purple'},
                      line_kws={'color': 'olive', 'linewidth': 1}
                      )
  chart.set_xlabel('Stress')
  chart.set_ylabel('Wellbeing')
  plt.title('Scatterplot of Wellbeing vs. Stress with Line of Best Fit')

  plt.show()


# execute procedure from main context
show_scatterplot(df)
````

## **About this**

### **Description:**
A scatterplot plots the relationship between stress levels and well-being scores for a group of individuals. Each point on the graph represents a pair of data points from the data set, where the x-axis represents stress levels and the y-axis represents well-being scores. The line of best fit, which is generated by linear regression, represents the overall trend in the data.

### **Conclusions:**
From the scatterplot and the line of best fit, we can observe the following:

#### 1. *Trend Observation:*
- The line of best fit indicates whether there is a positive or negative relationship between stress and well-being.
- In our case, the line slopes downward from left to right, suggesting that higher stress levels are associated with lower well-being scores, implying a negative relationship.

#### 2. *Confidence in Conclusion:*
- The strength of the relationship can be assessed by examining how closely the data points cluster around the line of best fit.
- Tight clustering around the line suggests a stronger relationship and greater confidence in the observed trend. But this is not our case, as we can see.
- Our data points are widely scattered and the line does not fit, the relationship may be weak or there may be other factors influencing the well-being indicators.

### ***To summarize***
The scatterplot and the line of best fit give an idea of ​​the correlation between stress and well-being. By analyzing the trend indicated by the line and the distribution of the data points, we can draw conclusions about the nature and strength of the relationship between these two variables.

### Exercise 8 - Create a set of charts from a data set
---
Use the student exam scores dataset here: https://raw.githubusercontent.com/lilaceri/Working-with-data-/main/Data%20Sets%20for%20code%20divisio/student_scores.csv

Investigate the data and create a set of charts.  

Create each chart in a new code cell.

Add a text cell below each visualisation to explain what the chart is showing.


````python
# Create a set of charts from a data set
def show_charts(df):
  # 1. Scatter Plot with Line of Best Fit
  plt.figure(figsize=(10, 6))
  chart = sns.regplot(x='Hours', y='Scores', data=df,
                      scatter_kws={'color': 'navy', 's': 100},
                      line_kws={'color': 'salmon', 'linewidth': 2})

  chart.set_xlabel('Hours', fontsize=12)
  chart.set_ylabel('Scores', fontsize=12)
  plt.title('Scatter Plot of Hours vs. Scores with Line of Best Fit', fontsize=14)

  plt.show()
  print('''
  This scatter plot shows the relationship between the number of hours studied
  and the scores achieved. The red line represents the line of best fit,
  which helps to visualize any correlation. If the points are
  closely clustered around the line, it indicates a strong relationship
  between hours studied and scores
  ''')

  #2. Histogram of Hours
  plt.figure(figsize=(10, 6))
  chart = sns.histplot(df['Hours'], bins=10, color='green', kde=False)

  chart.set_xlabel('Hours', fontsize=12)
  chart.set_ylabel('Frequency', fontsize=12)
  plt.title('Histogram of Hours', fontsize=14)

  plt.show()
  print('''
  The histogram displays the distribution of study hours among individuals.
  It shows how frequently different ranges of hours are observed.
  This helps in understanding the concentration of data
  in specific ranges of hours
  ''')

  #3. Histogram of Scores
  plt.figure(figsize=(10, 6))
  chart = sns.histplot(df['Scores'], bins=10, color='purple', kde=False)

  chart.set_xlabel('Scores', fontsize=12)
  chart.set_ylabel('Frequency', fontsize=12)
  plt.title('Histogram of Scores', fontsize=14)

  plt.show()
  print('''
  This histogram provides insights into the distribution of scores.
  It shows which score ranges are most common and can help in identifying
  patterns or outliers in the scores
  ''')

  # 4. Box Plot of Scores
  plt.figure(figsize=(10, 6))
  chart = sns.boxplot(y=df['Scores'], color='orange')

  chart.set_ylabel('Scores', fontsize=12)
  plt.title('Box Plot of Scores', fontsize=14)

  plt.show()
  print('''
  The box plot summarizes the distribution of scores, showing the median,
  quartiles, and potential outliers. It provides a clear view of the spread
  and central tendency of the scores
  ''')

  # 5. Box Plot of Hours
  plt.figure(figsize=(10, 6))
  chart = sns.boxplot(y=df['Hours'], color='cyan')

  chart.set_ylabel('Hours', fontsize=12)
  plt.title('Box Plot of Hours', fontsize=14)

  plt.show()
  print('''
  This box plot provides a summary of the hours data,
  highlighting the median, quartiles, and any potential outliers.
  It helps in understanding the spread and central tendency
  of the study hours
  ''')

  # 6. Histogram of Hours with KDE Line
  plt.figure(figsize=(10, 6))
  chart = sns.histplot(df['Hours'], bins=10, color='green', kde=True)

  chart.set_xlabel('Hours', fontsize=12)
  chart.set_ylabel('Frequency', fontsize=12)
  plt.title('Histogram of Hours with KDE Line', fontsize=14)

  plt.show()
  print('''
  The histogram with the KDE line provides a visual representation
  of the data distribution for Hours, with the KDE line
  offering a smoothed estimate of the probability density.
  This helps to understand the underlying distribution shape more clearly
  ''')

  # 7. Histogram of Scores with KDE Line
  plt.figure(figsize=(10, 6))
  chart = sns.histplot(df['Scores'], bins=10, color='purple', kde=True)

  chart.set_xlabel('Scores', fontsize=12)
  chart.set_ylabel('Frequency', fontsize=12)
  plt.title('Histogram of Scores with KDE Line', fontsize=14)

  plt.show()
  print('''
  Similarly, the histogram with the KDE line for Scores
  provides a visual representation of the data distribution,
  with the KDE line helping to smooth out the histogram bars
  and giving a clearer view of the data's density and distribution
  ''')

  # 8. Line Plot of Scores Over Hours
  plt.figure(figsize=(10, 6))
  sns.lineplot(x='Hours', y='Scores', data=df, marker='o', color='blue')

  plt.xlabel('Hours', fontsize=12)
  plt.ylabel('Scores', fontsize=12)
  plt.title('Line Plot of Scores Over Hours', fontsize=14)
  plt.grid(True)

  plt.show()
  print('''
  This line plot shows how Scores change as the number of Hours increases.
  The markers on the line indicate individual data points.
  This plot helps to visualize any trends or patterns in the relationship
  between hours studied and scores achieved.
  ''')

  # 9. Line Plot of Moving Average of Scores
  # Calculate the moving average of Scores with a window size of 3
  df['Scores_MA'] = df['Scores'].rolling(window=3).mean()

  plt.figure(figsize=(10, 6))
  sns.lineplot(x='Hours', y='Scores', data=df, marker='o', label='Scores', color='blue')
  sns.lineplot(x='Hours', y='Scores_MA', data=df, label='Moving Average', color='red', linestyle='--')

  plt.xlabel('Hours', fontsize=12)
  plt.ylabel('Scores', fontsize=12)
  plt.title('Line Plot of Scores with Moving Average', fontsize=14)
  plt.legend()
  plt.grid(True)

  plt.show()
  print('''
  This plot shows both the actual Scores and a smoothed trend line (moving average).
  The moving average line (in red) helps to identify the underlying trend
  by smoothing out the volatility in scores
  ''')

  # 10. Pie Chart of Average Scores by Hours Range
  # Define bins and labels for hours
  bins = [0, 3, 6, 9, 12]
  labels = ['0-3 hours', '4-6 hours', '7-9 hours', '10-12 hours']
  df['Hours_Binned'] = pd.cut(df['Hours'], bins=bins, labels=labels)

  # Calculate the average score for each range
  average_scores = df.groupby('Hours_Binned')['Scores'].mean()

  # Plot pie chart
  plt.figure(figsize=(8, 8))
  colors = sns.color_palette('pastel')
  plt.pie(average_scores, labels=average_scores.index, colors=colors, autopct='%1.1f%%', startangle=140)

  plt.title('Pie Chart of Average Scores by Hours Range', fontsize=14)
  plt.show()
  print('''
  This pie chart shows the average scores within different
  ranges of study hours. It helps to visualize how scores
  are distributed across various time intervals
  and can reveal patterns related to study time
  ''')

  # 11. Pie Chart of Proportions of Scores Above and Below the Median
  # Calculate the median score
  median_score = df['Scores'].median()

  # Create categories based on median score
  df['Score_Category'] = df['Scores'].apply(lambda x: 'Above Median' if x > median_score else 'Below Median')

  # Count the number of scores in each category
  score_category_counts = df['Score_Category'].value_counts()

  # Plot pie chart
  plt.figure(figsize=(8, 8))
  colors = sns.color_palette('pastel')
  plt.pie(score_category_counts, labels=score_category_counts.index, colors=colors, autopct='%1.1f%%', startangle=140)

  plt.title('Pie Chart of Scores Above and Below the Median', fontsize=14)
  plt.show()
  print('''
  This pie chart shows the distribution of scores above and below the median.
  It provides insight into how scores are spread in relation to the median,
  which can be useful for understanding score distribution relative to a central value
  ''')


# execute procedure from main context
url = "https://raw.githubusercontent.com/lilaceri/Working-with-data-/main/Data%20Sets%20for%20code%20divisio/student_scores.csv"
df = get_csv_data(url)
show_charts(df)
````
